``````python
# Complete the word_frequency() function to count word frequencies sorted alphabetically.
#  Run the script on 'alice_in_wonderland.txt'.

def word_frequency(file):
  # Read the file
  with open(file, 'r') as f:
    text = f.read()
  
  # Remove punctuation
  for char in '-.,\n:\'!()"[]*;?_':
    text = text.replace(char, ' ')

  # Split the text into words
  words = text.lower().split()

  # Count the frequency of each word
  word_freq = {}
  for word in words:
    if word in word_freq:
      word_freq[word] += 1
    else:
      word_freq[word] = 1
  
  return word_freq

word_freq_asdfojas = word_frequency('alice_in_wonderland.txt')

# Sort the word frequencies alphabetically
word_freq_sorted = dict(sorted(word_freq_asdfojas.items()))

# Display first five word frequencies
for word, freq in list(word_freq_sorted.items())[:5]:
  print(f"{word}: {freq}")
``````

1. **Fibonacci Sequence**: Create a program that generates Fibonacci numbers less than a limit and writes them to a file. The _Fibonacci_ sequence is a sequence in which each number is the sum of the two preceding ones: 
	`0, 1, 1 (0+1), 2 (1+1), 3 (2+1), 5 (3+2), ...`
	- Use a function to generate Fibonacci numbers as a list
	- Use `with` statements for file operations
	- Handle potential file I/O errors with `try`/`except`
	- Use command-line arguments (via `argparse`) to specify the upper limit and output file name
	- **Task: Generate the Fibonacci numbers less than 100 and write them to `fibonacci_100.txt`


``````python
# Find all fibonacci numbers less than n
def fibonacci(n):
    fib = [1, 1]
    while True:
        next_fib = fib[-1] + fib[-2]
        if next_fib >= n:
            break
        fib.append(next_fib)
    return fib
    
display(fibonacci(100))
``````

``````python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

fib = fibonacci(50000)

# Find largest prime fibonacci number
for n in reversed(fib):
    if is_prime(n):
        display(n)
        break
``````

``````python
import random

# Create a file of random DNA sequences in FASTA format, 80 characters per line
def generate_fasta(length, file):
    # Generate random DNA sequence
    dna = ''.join(random.choices('ACGT', k=length))
    
    # Write to file
    with open(file, 'w') as f:
        f.write(f">random_sequence\n")
        for i in range(0, length, 80):
            f.write(dna[i:i+80] + '\n')

generate_fasta(1000000, 'random_dna.fasta')

!head random_dna.fasta
``````

## Question 4: Find Distant Cutsites in FASTA Data (30 points)
File name: `find_cutsites.py`

In molecular biology, restriction enzymes cut DNA at specific sequences called restriction sites or cut sites. Finding pairs of cut sites that are a certain distance apart is important for various genetic engineering techniques. Cutsites are often represented with a vertical bar (|) in the cut site sequence, indicating where the enzyme cuts the DNA.

An example:
- Take cut site sequence "G|GATCC" for BamHI:
- In the sequence dna="AAGG|GATCCTT", the cut site starts at index 4

Create a Python script that finds pairs of restriction enzyme cut sites that are 80-120 kilobase pairs (kbp) apart in a given FASTA file. Your script should:

``````python
# Find pairs of restriction enzyme cut sites that are 80-120 kilobase pairs (kbp) apart in a given FASTA file.
# 1. Accept two arguments: the FASTA file path (data/random_sequence.fasta) and a cut site sequence (e.g., "G|GATCC")
# 2. Read the FASTA file and save the DNA sequence to a variable omitting whitespace.
# 3. Find all occurrences of the cut site (specified below) in the DNA sequence.
# 4. Find all pairs of cut site locations that are 80,000-120,000 base pairs (80-120 kbp) apart.
# 5. Print the total number of cut site pairs found and the positions of the first 5 pairs.
# 6. Save a summary of the results (example below) in the results directory as "cutsite_summary.txt".

import re
import os

def find_cut_sites(file, cut_site):
    """Find all occurrences of a restriction enzyme cut site in a FASTA sequence.
    
    Args:
        file (str): Path to FASTA file
        cut_site (str): Cut site sequence with '|' marking cut position (e.g. "G|GATCC")
    
    Returns:
        list: Positions of all cut sites found
    """
    # Find the cut marker '|'
    offset = cut_site.index('|')

    # Get the recognition sequence without the cut marker
    cut_seq = cut_site.replace('|', '')
    
    # Read the FASTA file
    with open(file, 'r') as f:
        lines = f.readlines()
        if not lines or not lines[0].startswith('>'):
            raise ValueError("Invalid FASTA format - must start with '>'")
        # Join all lines except header, removing whitespace
        seq = ''.join(line.strip() for line in lines[1:])
    
    # Find all occurrences of the cut site (case insensitive)
    return [m.start() for m in re.finditer(cut_seq, seq, re.IGNORECASE)]

def find_pairs(positions, min_dist, max_dist):
    """Find pairs of positions that are within the specified distance range.
    
    Args:
        positions (list): List of integer positions
        min_dist (int): Minimum distance between pairs
        max_dist (int): Maximum distance between pairs
    
    Returns:
        list: List of tuples containing pairs of positions within range
    """
    pairs = []
    for i, pos1 in enumerate(positions):
        for pos2 in positions[i+1:]:
            if min_dist <= pos2 - pos1 <= max_dist:
                pairs.append((pos1, pos2))
    return pairs

# Find all cut sites
cut_sites = find_cut_sites('random_dna.fasta', 'G|GATCC')
print(f"Found {len(cut_sites)} cut sites")

# Find pairs 80-120 kb apart
pairs = find_pairs(cut_sites, 80000, 120000)
print(f"\nFound {len(pairs)} pairs 80-120 kb apart: {pairs[:5]}")

# Save results
os.makedirs('results', exist_ok=True)
with open('results/cutsite_summary.txt', 'w') as f:
    f.write(f"Cut sites found: {cut_sites}\n")
    f.write(f"Number of pairs 80-120 kb apart: {len(pairs)}\n")
    if pairs:
        f.write(f"First 5 pairs: {pairs[:5]}\n")
``````
